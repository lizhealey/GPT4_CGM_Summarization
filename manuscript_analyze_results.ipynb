```python
data_all = pd.DataFrame()

# Get statistics
for name in ['healthy_adult#008.csv','healthy_adult#006.csv','healthy_adult#007.csv',
             'dinner_adult#004.csv','dinner_adult#006.csv','dinner_adult#007.csv','mild_adult#006.csv',
    'breakfast_week_adult#005.csv','breakfast_week_adult#003.csv','breakfast_week_adult#008.csv']:
    name = 'date_corrected'+str(name)
    data=pd.read_csv(f'/Users/elizabeth/Documents/GPT4_CGM_Manuscript/processed_sim_data/{name}')
    
    name_ = name
    data_all.loc[name_,'Average']  = np.mean(data['CGM'])
    data_all.loc[name_,'GMI'] = 3.31 + 0.02392 *np.mean(data['CGM'])
    data_all.loc[name_,'CV' ]= np.std(data['CGM'])/np.mean(data['CGM'])
    data_all.loc[name_,'Mininum'] = np.min(data['CGM'])
    data_all.loc[name_,'Maximum'] = np.max(data['CGM'])
    data_all.loc[name_,'Percent active'] = len(data['CGM'])/(14*288)
    tbh= data[data['CGM']<180]
    data_all.loc[name_,'TIR'] = len(tbh[tbh['CGM']>70])/(len(data['CGM']))
    data_all.loc[name_,'TAR 1 (>180)'] = len(data[data['CGM']>180])/(len(data['CGM']))
    data_all.loc[name_,'TAR 2 (>300)'] = len(data[data['CGM']>250])/(len(data['CGM']))
    data_all.loc[name_,'TBR 1 (<70)'] = len(data[data['CGM']<70])/(len(data['CGM']))
    data_all.loc[name_,'TBR 2 (<54)'] = len(data[data['CGM']<54])/(len(data['CGM']))
    
    
```
